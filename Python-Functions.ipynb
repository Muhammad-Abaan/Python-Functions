1. What is the difference between a function and a method in Python?
    - A function is a block of reusable code that performs a specific task.
    - A method is essentially a member function of a class. It's defined inside a class and is designed to operate on instances of that class or the class itself.
2. Explain the concept of function arguments and parameters in Python.
    - parameters are basically variables in the function definition. "def greet(name):  # 'name' is a parameter"
    - arguments are values passed during the function call. "greet("Alice")  # 'Alice' is an argument."
3. What are the different ways to define and call a function in Python?
    - we can define functions using def, lambda, or within other functions
      and we can call them by their name, passing arguments, or return them from other functions.
4. What is the purpose of the return statement in a Python function?
    - The return statement in a python function is used to send a result or value back to the caller. It marks the end of the function's execution and specifies what the function should output when it is called.
5. What are iterators in Python and how do they differ from iterables?
    - An iterable is any python object/sequential structure/data structure that is capable of returning its members one at a time, permitting it to be iterated over in a for loop, example: list, tuples
    - an iterator is an object representing a stream of data; return the data one by one.
6. Explain the concept of generators in Python and how they are defined.
    - Generators are a way to create iterators using a function. They produce values one at a time as they are needed, instead of computing and storing them all at once.
    - Generators use the yield keyword instead of return. When yield is encountered, the function's state is "paused," and the value is sent to the caller. Execution resumes from the paused state the next time the generator is called.
    - yeild may return an iterator object, which resumes and return the value when next() is used to fetch next value.
7. What are the advantages of using generators over regular functions?
    - Generators generate items one at a time and do not store the entire result in memory. Regular functions often return complete data structures (like lists), which can consume significant memory if the dataset is large.
    - Also, for streaming data or real-time processing, generators can provide advantage over regular functions.
8. What is a lambda function in Python and when is it typically used?
    - A lambda function in Python is a small, anonymous function defined using the lambda keyword. Unlike regular functions defined with def, lambda functions can have only one expression, which is evaluated and returned automatically.
    - They can have only a single expression and cannot contain multiple statements or a return statement. Lambda functions also do not support default arguments.
    - they are basically shorthand for a single expression function.
    - used for simple, short-lived operations.
9. Explain the purpose and usage of the map() function in Python.
    - The map() function in Python is a built-in function used to apply a given function to all items in an iterable (like a list, tuple, or any other collection). The result is a map object, which can be easily converted to a list or another iterable.
    - The main purpose of map() is to transform or apply a function to each element in an iterable without using explicit loops.
10. What is the difference between map(), reduce(), and filter()  functions in Python?
    - map() is used when you need to transform the elements of an iterable, like applying a function to all elements.
    - reduce() is used when you need to reduce the iterable to a single value, such as when calculating the sum or product of all elements.
    - filter() is used when you need to filter out elements based on a condition, like extracting elements that satisfy a specific condition.

```python
# 1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in the list.
def sum(l):
  sum = 0
  for i in l:
    if i%2==0:
      sum += i
  return sum
l = list(map(int, (input("Enter the numbers: ").split())))
print(sum(l))
```

```python
# 2. Create a Python function that accepts a string and returns the reverse of that string.
def rev(s):
  return s[::-1]
s = input("Enter the string: ")
print(rev(s))
```

```python
# 3. Implement a Python function that takes a list of integers and returns a new list containing the squares of each number.
def sqr(l):
  a = []
  for i in l:
    a.append(i**2)
  return a
l = list(map(int, input("Enter the numbers: ").split()))
print(sqr(l))
#       OR
print([i**2 for i in l])
```

```python
# 4. Write a Python function that checks if a given number is prime or not from 1 to 200.
def prime(n):
  i = 2
  b = True
  if n <= 1:
    return False
  while i <= n**0.5:
    if n%i == 0:
      b = False
      break
    i +=1
  if b == True:
    return True
  else:
    return False
n = int(input("Enter the number: "))
print(prime(n))
```

```python
# 5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of terms.
def fib(n):
  a = 0
  b = 1
  for i in range(n):
    yield a
    a, b = b, a+b
n = int(input("Enter the number: "))
a = (fib(n))
for i in range(n):
  print(next(a))
```

```python
# 6. Write a generator function in Python that yields the powers of 2 up to a given exponent.
def exp(n):
  for i in range(0,n+1):
    yield f"2**{i} = {2**i}"
n = int(input("Enter the number: "))
a = exp(n)
for i in range(n):
  print(next(a))
```

```python
# 7. Implement a generator function that reads a file line by line and yields each line as a string.
# Method 1 (General Basic Code)
def line(s):
  sentence = s.split("\n")
  for i in sentence:
    yield i.strip()
s = input("Enter the text: ")
for i in line(s):
  print(i)

# Method 2 (Actual Code for reading a text file provided by path )
# def read(file_path):
#   with open(file_path, 'r') as file:               # file is basically iterator object of .txt provided through a path opened in read only mode
#     for line in file:
#       l = line.split(".")  # although the line would iterate in file at \n intervals, i aslo included split at full stop to filter it out as a sencetence
#       for i in l:
#         yield i.strip()

# file_path = input("Enter the file path: ")
# try:
#   for i  in read(file_path):
#     print(i)
# except FileNotFoundError:
#   print("The file does not exist.")
```

```python
# 8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.
l = [(1, 4), (3, 1), (5, 2), (2, 3)]
print(sorted(l, key=lambda x : x[1]))
```

```python
# 9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit.
c = list(map(int, input("Enter the temperatures in celcius: ").split()))
f = list(map(lambda x : (x*(9/5))+32, c))
print(f)
```

```python
# 10.  Create a Python program that uses `filter()` to remove all the vowels from a given string.
s = input("Enter the string: ")
def vow(s):
    vowels = "aeiouAEIOU"
    return s not in vowels
a = ''.join((filter(vow, s)))
print(a)
```

```python
"""11.  Imagine an accounting routine used in a book shop. It works on a list with sublists,
Write a Python program, which returns a list with tuples. Each tuple consists of the order number and the
price per item and the quantity. The product quantity should be increased by 10, if the value of the
order is smaller than 100,00 .
Write a Python program using lambda and map."""
s = input("Enter the values as specified in question: ")  # [order_number, price_per_item, quantity]
l = eval(s)
inc = lambda x : (x[0], x[1], x[2]+10) if (x[1]*x[2]) < 10000 else (x[0], x[1], x[2])
ans = list(map(inc, l))
print(ans)
```
